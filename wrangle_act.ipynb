### Issue #6: Create stage column and collapse doggo, floofer, pupper, puppo into stage column

#### Define

- Here, I created a function that creates the stage column and writes the value of each row from doggo, floofer, pupper and puppo into that stage column. Then i dropped the columns since they it has been recorded in the stage column

#### Code

```python
def dog_stage(row):
    dog_stage_list = []
    # Setting the conditional
    if row['doggo'] == 'doggo':
        # Adding the entry to the list if it satisfies the condition
        dog_stage_list.append('doggo')
    if row['floofer'] == 'floofer':
        dog_stage_list.append('floofer')
    if row['pupper'] == 'pupper':
        dog_stage_list.append('pupper')
    if row['puppo'] == 'puppo':
        dog_stage_list.append('puppo')
    if len(dog_stage_list) < 1: 
        row['stage'] = 'None'
    else: 
        row['stage'] = ' and '.join(dog_stage_list)
    return row
```

```python
# Apply the function to the table
dog_data_clean = dog_data_clean.apply(dog_stage, axis=1)
```

```python
# Drop the columns after applying the function
dog_data_clean = dog_data_clean.drop(['doggo','floofer','pupper','puppo'], axis =1)
```

### Write a function to remove outliers

#### Code

```python
def remove_outliers(dataframe, column):
    '''
    the function take a dataframe and a column as argument 
    cut the outlier of the dataframe
    and return the new dataframe shape
    '''
    q3 = dataframe[column].quantile(0.75)
    q1 = dataframe[column].quantile(0.25)
    IQR = q3 - q1
    outlier_1 = q1 - 1.5*IQR
    outlier_2 = q3 + 1.5*IQR
    dataframe = dataframe.loc[((dataframe[column] > outlier_1) & (dataframe[column] < outlier_2))]
    return dataframe
```

```python
# Apply function to 'reweet_count' column
tweet_data_clean = remove_outliers(tweet_data_clean,'retweet_count')
```

```python
# Apply function to 'rating' column
tweet_data_clean = remove_outliers(tweet_data_clean,'rating')
```
